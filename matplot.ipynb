```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x** 2
```

```python
plt.plot(x,y,'r-.')
plt.plot(x+5,y+5,'k-.')
plt.xlim(-10,20)
plt.ylim(-10,20)
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'c*-')
plt.subplot(1,3,2)
plt.plot(x,y,'b+-')
plt.subplot(1,3,3)
plt.plot(x,y,'r*-')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'c+-')
plt.title("graph")
plt.xlabel("v")
plt.ylabel("s")
plt.plot(x,y,'m*-',label='BCA')
plt.legend()



plt.subplot(2,2,2)
plt.plot(x,y,'r*-')
plt.title("not graph")
plt.xlabel("v")
plt.ylabel("s")
plt.plot(x,y,'g*-',label='CSE')
plt.legend()


plt.subplot(2,2,3)
plt.plot(x,y,'m*-')
plt.title("not graph")
plt.xlabel("v")
plt.ylabel("s")
plt.plot(x,y,'r*-',label='DSC')
plt.legend()



plt.subplot(2,2,4)
plt.plot(x,y,'k^-')
plt.title("graph")
plt.xlabel("v")
plt.ylabel("s")
plt.plot(x,y,'k*-',label='BBA')
plt.legend()
plt.tight_layout()
```

```python
x = np.linspace(0,6,11)
y = x**2

fig = plt.figure()

axes = fig.add_axes([1,1,3,3])

axes.plot(x,y,'b-.',label='axes')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('Ax1')
axes.legend()


axes = fig.add_axes([1.2,2.5,1,1])
axes.plot(x,y,'r.-',label='axes')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('Ax2')
axes.legend()


```
